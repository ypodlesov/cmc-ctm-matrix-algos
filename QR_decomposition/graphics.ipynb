```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
# import numpy as np

mult = pd.read_csv("./benchmarks/multiply.csv")
block_mult = mult[mult["name"].str.startswith("BM_BlockProd")]
block_mult.loc[:, "name"] = block_mult["name"].str.replace('BM_BlockProd/', '')
simple_mult = mult[mult["name"].str.startswith("BM_Prod")]
simple_mult.loc[:, "name"] = simple_mult["name"].str.replace('BM_Prod/', '')

plt.figure(figsize=(10, 6))
plt.plot(block_mult["name"], block_mult["real_time"] * 1e-9)
plt.plot(simple_mult["name"], simple_mult["real_time"] * 1e-9)
plt.legend(['block_mult', 'simple_mult'])
plt.title('Время выполнения умножения матрицы от порядка матрицы')
plt.xlabel('Порядок матрицы')
plt.ylabel('Real time (s)')
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
# import numpy as np

mult = pd.read_csv("./benchmarks/qr_decomposition.csv")
simple_qr = mult[mult["name"].str.startswith("BM_QRDecomposition/")]
simple_qr.loc[:, "name"] = simple_qr["name"].str.replace('BM_QRDecomposition/', '')

block_qr = mult[mult["name"].str.startswith("BM_QRDecompositionBlockOptimized")]
block_qr.loc[:, "name"] = block_qr["name"].str.replace('BM_QRDecompositionBlockOptimized/', '')

plt.figure(figsize=(10, 6))
plt.plot(simple_qr["name"], simple_qr["real_time"] * 1e-9)
plt.plot(block_qr["name"], block_qr["real_time"] * 1e-9)
plt.legend(['simple_qr', 'block_optimized_qr'])
plt.title('Время выполнения QR разложения матрицы от порядка матрицы')
plt.xlabel('Порядок матрицы')
plt.ylabel('Real time (s)')
plt.show()
```
