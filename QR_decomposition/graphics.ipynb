## График

```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv("./build/result.csv")
data["name"] = data["name"].str.replace('BM_QRDecomposition/', '')
data["real_time"] = data["real_time"] * 1e-6

plt.figure(figsize=(10, 6))
plt.plot(data["name"], data["real_time"])
plt.title('Время выполнения алгоритма QR разложения методом Грама Шмидта от порядка матрицы')
plt.xlabel('Порядок матрицы')
plt.ylabel('Real time (ms)')
plt.show()
```
